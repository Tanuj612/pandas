```python
# column selectionüòä
import pandas as pd
d={'one':pd.Series([1,2,3,],index=['a','b','c']),'two':pd.Series([1,2,3,4],index=['a','b','c','d'])}
df=pd.DataFrame(d)
print(df)
print(df['one'])
```

```python
# column addition and sum of the given columns üëç
import pandas as pd
d={'one':pd.Series([1,2,3,],index=['a','b','c']),'two':pd.Series([1,2,3,4],index=['a','b','c','d'])}
df=pd.DataFrame(d)
df['three']=pd.Series([10,20,30],index=['a','b','c'])
df['four']=df['one']+df['three']
print(df)
```

```python
# for delete the column using del function and pop methodüòÅ
import pandas as pd
d={'one':pd.Series([1,2,3,],index=['a','b','c']),'two':pd.Series([1,2,3,4],index=['a','b','c','d']),'three':pd.Series([10,20,30],index=['a','b','c'])}
df=pd.DataFrame(d)
print(df)
del df['one']
print(df)
df.pop('two')
print(df)
```

```python
# for transpose üôÉ
import pandas as pd
d={'name':pd.Series(['tom','james','ricky']),'age':pd.Series([23,26,27]),'rating':pd.Series([4.23,4.6,3.8])}
df=pd.DataFrame(d)
print(df.T)
```
